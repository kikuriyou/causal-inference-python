```python
import os
from time import time
import math
import random
import gc
import warnings

import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from lifelines.datasets import load_rossi

warnings.filterwarnings('ignore')
np.random.seed(123)
pd.set_option('display.max_columns', 100)
plt.style.use('seaborn-dark')

save_dir = 'sample_telco'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
csv_path = '../input/WA_Fn-UseC_-Telco-Customer-Churn.csv'

print("Loading dataset...")
t0 = time()
df = pd.read_csv(csv_path)
print(f'End in {time() - t0 :.2f}s.')

print(f'shape: {df.shape}')
display(df.head(2))
```

# Check dataset

```python
print(df.columns)
#for col in df.columns:
#    print(f'{col}, {df[col].nunique()}')
```

```python
treat_col = 'Partner'
ps_col = 'propensity'
outcome_col = 'event_time'
```

# Prepare event time, preprocess covariates

```python
max_time = df['tenure'].max()
df['event_time'] = np.where(df['Churn']=='Yes', df['tenure'], max_time)

drop_cols = ['tenure', 'Churn']
df = df.drop(drop_cols, axis=1)

df['Partner'] = np.where(df['Partner']=='Yes', 1, 0)

df['TotalCharges'] = df['TotalCharges'].replace(' ', 0).astype('float64')
df['MonthlyCharges'] = df['MonthlyCharges'].astype('float64')
```

```python
from sklearn.preprocessing import MinMaxScaler
cols = ['TotalCharges', 'MonthlyCharges']
scaler = MinMaxScaler()
df.loc[:, cols] = scaler.fit_transform(df[cols].values)
```

```python
# inc, pmoney, age: categorical -> one-hot encoding
biased_df = df.copy()
cat_cols = ['gender', 'Dependents', 'PhoneService', 'MultipleLines', 
            'InternetService', 'OnlineSecurity', 'OnlineBackup', 
            'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
            'Contract', 'PaperlessBilling', 'PaymentMethod']
for col in cat_cols:
    dummy_df = pd.get_dummies(biased_df[col], prefix=col, drop_first=True)
    biased_df = pd.concat([biased_df, dummy_df], axis=1)
    biased_df = biased_df.drop(col, axis=1)
print(biased_df.columns)
biased_df.head(2)
```

```python
cnt_tr = len(biased_df[biased_df[treat_col]==1])
cnt_ct = len(biased_df[biased_df[treat_col]==0])

print(f'# of samples (treatment): {cnt_tr}')
print(f'# of samples (control):   {cnt_ct}')
```

# Estimate biased effects

```python
from causal_inference import show_survival_curve
save_file = os.path.join(save_dir, 'survival_curve_biased.png')
biased_ate = show_survival_curve(biased_df, t_col=treat_col, y_col=outcome_col, save_file=save_file)
```

# Estimate effects with adjusting by propensity score

```python
from causal_inference import EffectEstimatorPS, show_covariate_distribution
```

```python
# drop target-like variables, and drop_first in existing dummy variables
drop_cols = ['customerID']
cov_cols = [col for col in biased_df.columns if col not in [treat_col, ps_col, outcome_col] + drop_cols]
print(f'# of covariates: {len(cov_cols)}')
print(cov_cols)
```

```python
# Check distribution
save_file = os.path.join(save_dir, 'covariate_distribution.png')
show_covariate_distribution(biased_df, t_col=treat_col, cov_cols=cov_cols, save_file=save_file)
```

# IPW
---

```python
# weighting
est = EffectEstimatorPS(t_col=treat_col, cov_cols=cov_cols, ps_col=ps_col, y_col=outcome_col, ps_clipping=True, 
                        method='weighting', weighting_method='ate', estimand='survival')
```

```python
# Estimate propensity score
ps_model = LogisticRegression(random_state=123)
biased_df[ps_col] = est.estimate_ps(model=ps_model, df=biased_df)

save_file = os.path.join(save_dir, 'ps_distribution.png')
est.show_ps_distribution(biased_df, title='logistic regression', save_file=save_file)
```

```python
weighted_df = est.adjust(biased_df)

save_file = os.path.join(save_dir, 'covariate_balance_weighting.png')
est.show_covariate_balance(unadjusted_df=biased_df, adjusted_df=weighted_df, save_file=save_file)    # , verbose=True
```

```python
save_file = os.path.join(save_dir, 'survival_curve_weighted.png')
est.estimate(weighted_df, save_file=save_file)
```

```python
elapsed_days = 10
print(f'ATE(weighting) after {elapsed_days} days: {est.effect_[elapsed_days] :.6f}')
```

## Matching
---

```python
est = EffectEstimatorPS(t_col=treat_col, cov_cols=cov_cols, ps_col=ps_col, y_col=outcome_col, 
                        method='matching', estimand='survival', caliper=1.e-1)
biased_df[ps_col] = est.estimate_ps(model=ps_model, df=biased_df)
```

```python
# Adjust confoundedness and check adjusted covariate balance
att_matched_df, atc_matched_df = est.adjust(biased_df)

print('\nChecking covariate balances....')
save_file = os.path.join(save_dir, 'covariate_balance_matching.png')
est.show_covariate_balance(unadjusted_df=biased_df, adjusted_df=att_matched_df, save_file=save_file)  # , verbose=True
#est.show_covariate_balance(unadjusted_df=biased_df, adjusted_df=atc_matched_df)  # , verbose=True
```

```python
# Estimate effect
print('\nATT results:')
save_file = os.path.join(save_dir, 'survival_curve_matched_att.png')
est.estimate(att_matched_df, save_file=save_file)
att = est.effect_

print('\nATC results:')
save_file = os.path.join(save_dir, 'survival_curve_matched_atc.png')
est.estimate(atc_matched_df, save_file=save_file)
atc = est.effect_
```

```python
elapsed_days = 10
print(f'ATT(matching) after {elapsed_days} days: {att[elapsed_days] :.6f}')
```
